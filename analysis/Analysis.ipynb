## Success rate for each strategy

````python
plt.figure(figsize=(12, 8))

# Setup columns
df_no_na = df[df['answer'].notna()]

# Success rate
success_rate_df = df_no_na.groupby(['strategy', 'depth'])['correct'].mean().reset_index()

for i, strategy in enumerate(success_rate_df['strategy'].unique()):
    strategy_data = success_rate_df[success_rate_df['strategy'] == strategy]

    strategy_data
    
    # Apply a rolling average with window size 3 (you can adjust this value)
    #rolling_mean = strategy_data['correct'].rolling(window=5, min_periods=1).mean()
    #plt.plot(strategy_data['depth'], rolling_mean, label=f"{strategy} (n={len(strategy_data)})", color=colors[i], marker=markers[i])
    
    plt.plot(strategy_data['depth'], strategy_data['correct'], label=strategy, color=colors[strategy], marker=markers[strategy])


# Set plot labels and title
plt.ylim(0, 1.1)
plt.xlabel('Depth')
plt.ylabel('Success Rate')
plt.title('Success Rate for Each Depth and Strategy (without "NA" answers)')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=4)
plt.grid(True)
plt.show()
````

## Success rate for YES/NO cases

````python
plt.figure(figsize=(12, 8))

# Extract 'yes' and 'no' cases from the 'prompt' column
df_no_na['case'] = df_no_na['promptFile'].apply(lambda x: 'yes' if 'yes' in x else 'no')

# Group by strategy, case, and depth, calculate success rate and count
grouped_df = df_no_na.groupby(['strategy', 'case', 'depth'])['correct'].agg(['mean', 'count']).reset_index()

# Plot the success rate for each strategy with different lines and colors for 'yes' and 'no' cases
plt.figure(figsize=(12, 8))

for i, strategy in enumerate(grouped_df['strategy'].unique()):
    for case in ['yes', 'no']:
        if case == 'yes':
            lineStyle = 'solid'
        else:
            lineStyle = 'dashed'
        case_data = grouped_df[(grouped_df['strategy'] == strategy) & (grouped_df['case'] == case)]
        
        # Apply a rolling average with window size 3 (you can adjust this value)
        rolling_mean = case_data['mean'].rolling(window=5, min_periods=1).mean()

        plt.plot(case_data['depth'], rolling_mean, label=f"{strategy} - {case} (n={case_data['count'].sum()})", marker=markers[strategy], color=colors[strategy], linestyle=lineStyle)

# Set plot labels and title
plt.ylim(0, 1.1)
plt.xlabel('Depth')
plt.ylabel('Success Rate')
plt.title('Success Rate for Each Depth, Strategy and Case (without "NA" answers)')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=4)
plt.grid(True)
plt.show()
````
